```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv(datalink)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
cluster_df = df.drop(['Name','Sex','Ticket','Cabin','Embarked'], axis=1)
cluster_df.head()
```

```python
cluster_df = cluster_df.dropna()
cluster = KMeans(n_clusters=3, random_state=0)
cluster.fit(cluster_df)
```

```python
cluster.fit_predict(cluster_df)
```

```python

cluster_labels = cluster.fit_predict(cluster_df)
df['cluster'] = np.nan
df.loc[cluster_df.index, 'cluster'] = cluster_labels
df['cluster'] = df['cluster'].astype('Int64')
df.head()
```

```python
df['cluster'].value_counts()
```

```python
df['cluster'].value_counts().hist()
```

```python
import seaborn as sns

sns.pairplot(df, hue='cluster')
```

```python
sns.boxplot(x=df['Fare'],y=df['cluster'])
```
